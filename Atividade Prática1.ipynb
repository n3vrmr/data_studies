```python
# Importando os dados
import pandas as pd

dados = pd.read_csv('ds_salaries.csv')
dados
```

```python
# Informações básicas
## Determine o número e nome das colunas do conjunto de dados
columns = dados.columns
print(columns)
print("Número de colunas:",len(columns))

## Determine o tipo de dado de cada coluna
print("Tipos de dados:",dados.dtypes)

## Determine o número de observações do conjunto de dados
print("Número de observações:",dados.count())

## Determine os valores máximos e mínimos das colunas numéricas
##e a amplitude desses dados
print("Coluna work_year:")
print("Máximo:",dados['work_year'].max())
print("Mínimo:",dados['work_year'].min())
print("Amplitude:",dados['work_year'].max()-dados['work_year'].min())

print("Coluna salary:")
print("Máximo:",dados['salary'].max())
print("Mínimo:",dados['salary'].min())
print("Amplitude:",dados['salary'].max()-dados['salary'].min())

print("Coluna salary:")
print("Máximo:",dados['salary'].max())
print("Mínimo:",dados['salary'].min())
print("Amplitude:",dados['salary'].max()-dados['salary'].min())

print("Coluna salary_in_usd:")
print("Máximo:",dados['salary_in_usd'].max())
print("Mínimo:",dados['salary_in_usd'].min())
print("Amplitude:",dados['salary_in_usd'].max()-dados['salary_in_usd'].min())

print("Coluna remote_ratio:")
print("Máximo:",dados['remote_ratio'].max())
print("Mínimo:",dados['remote_ratio'].min())
print("Amplitude:",dados['remote_ratio'].max()-dados['remote_ratio'].min())
```

```python
# Estatística descritiva
## Obtenha a média e mediana das colunas numéricas
print("Coluna work_year:")
print("Média:",dados['work_year'].mean())
print("Mediana:",dados['work_year'].median())

print("Coluna salary:")
print("Média:",dados['salary'].mean())
print("Mediana:",dados['salary'].median())

print("Coluna salary_in_usd:")
print("Média:",dados['salary_in_usd'].mean())
print("Mediana:",dados['salary_in_usd'].median())

print("Coluna remote_ratio:")
print("Média:",dados['remote_ratio'].mean())
print("Mediana:",dados['remote_ratio'].median())

## Obtenha o desvio-padrão
print("Coluna work_year:")
print("Desvio padrão:",dados['work_year'].std())

print("Coluna salary:")
print("Desvio padrão:",dados['salary'].std())

print("Coluna salary_in_usd:")
print("Desvio padrão:",dados['salary_in_usd'].std())

print("Coluna remote_ratio:")
print("Desvio padrão:",dados['remote_ratio'].std())
```

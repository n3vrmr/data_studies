```python
df.columns = ['data', 'cidade', 'faixa_etária', 'salário_residência',
              'falta_por_menstruação', 'motivo_falta', 'absorventes_na_escola',
              'acesso_absorventes_remédios', 'kits_essencial', 'itens_kit',
              'escola', 'comentário']
df
```

```python
idades = []
for item in df['faixa_etária']:
    if '12' in item:
        idades.append('12-14')
    elif '15' in item:
        idades.append('15-17')
    else:
        idades.append('>18')
df['faixa_etária'] = idades
df
```

# Importando os dados

```python
import pandas as pd

df = pd.read_excel('Informações de contato (respostas).xlsx')
df
```

# Limpeza e tratamento dos dados

```python
df.columns = ['data', 'cidade', 'faixa_etária', 'salário_residência',
              'falta_por_menstruação', 'motivo_falta', 'absorventes_na_escola',
              'acesso_absorventes_remédios', 'kits_essencial', 'itens_kit',
              'escola', 'comentário']
df
```

```python
idades = []
for item in df['faixa_etária']:
    if '12' in item:
        idades.append('12-14')
    elif '15' in item:
        idades.append('15-17')
    else:
        idades.append('>18')
df['faixa_etária'] = idades
df
```

```python
for item in df['salário_residência']:
    str(item)
df
```

```python
df.dtypes
```

```python
import numpy as np

salaries = []
for item in df['salário_residência']:
    if 'Mais de' in item:
        salaries.append('>3000')
    elif '2.500,00 e R$3' in item:
        salaries.append('2500-3000')
    elif '1.000,00 e R$2' in item:
        salaries.append('1000-2500')
    elif '500,00 e R$1' in item:
        salaries.append('500-1000')
    else:
        salaries.append(np.NaN)
df['salário_residência'] = salaries
df
```

```python
df.dtypes
```

```python
motivos = []
for item in df['motivo_falta']:
    mot = item.strip().split(sep=',')
    motivos.append(mot)
df['motivo_falta'] = motivos
df
```

```python
import pandas as pd

df = pd.read_excel("pda_especies_ameacadas.xlsx")
df
```

```python
df.columns = ["sq_ficha", "sq_taxon", "nome_científico",
              "nome_comum", "grupo_taxonômico", "categoria_risco",
              "unidade_conservação_ocorrência", "código_cnuc"]
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
categorias = df["categoria_risco"].value_counts()
categorias
```

```python
risk = ["Vulnerável (VU)", "Em Perigo (EN)",
        "Criticamente em Perigo (CR)",
        "Extinta na Natureza (EW)"]
```

```python
plt.pie(categorias, labels=risk, autopct='%1.2f%%')
```

```python
grupos_taxon = df["grupo_taxonômico"].value_counts()
grupos_taxon
```

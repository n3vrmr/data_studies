```python
import pandas as pd
import datetime
```

```python
# Importe os dados do arquivo Space_Corrected.csv como um DataFrame
space = pd.read_csv("Space_Corrected.csv")
space
```

```python
# Análise descritiva de dados
# Quantos elementos estão presentes no conjunto de dados?
print("Número de elementos:",len(space))
# Quantas colunas?
print("\nNúmero de colunas:",len(space.columns))
```

```python
# A coluna Datum possui informações sobre a data de lançamento
# Para transformar a data de texto para datetime e poder obter o ano, vamos utilizar a função to_datetime
# Assim vamos obter o ano de cada lançamento e salvar em uma nova coluna
space['Ano'] = [i.year for i in pd.to_datetime(space['Datum'])]
space.head()
```

```python
# Obtenha quantos lançamentos foram realizados por ano e qual ano teve mais lançamentos
ano = space["Ano"].value_counts()
print(ano)
print("\nAno com mais lançamentos: 1971, com 119 lançamentos.")
# Esse conjunto de dados compreende quantos anos de lançamento?
print(f"\nEsse conjunto compreende {len(ano)} anos de lançamentos.")
```

```python
# Determine quantos lançamentos foram bem sucedidos
success = space[space["Status Mission"] == "Success"]
print(f"Foram {len(success)} lançamentos bem sucedidos.")
```

```python
# Determine quantos lançamentos foram realizados por cada Company Name
company = space.groupby("Company Name").count()
print("Lançamentos por empresa:",company)
# Quantos lançamentos são brasileiros?
brasil = []
for country in space["Location"]:
    if "Brazil" in country:
        brasil.append(country)
print("\nLançamentos realizados no Brasil:",len(brasil))
```

```python
# Obtenha todos os lançamentos realizados em 1969
moon = space[space["Ano"] == 1969]
print(moon)
# Quantos foram realizados pelos NASA? E pela RVSN USSR?
nasa = space[space["Company Name"] == "NASA"]
ussr = space[space["Company Name"] == "RVSN USSR"]
print("\nLançamentos da NASA:",len(nasa),
      "\nLançamentos da USSR:",len(ussr))
# Um dos lançamentos de 1969 foi o da missão até a lua.
# Essa missão foi realizada em um foguete Saturn V | Apollo 11, o que estará na coluna Detail
# Encontre esse lançamento no conjunto de dados e determine a data desse lançamento
neil = space[space["Detail"] == "Saturn V | Apollo 11"]
print("\n\nData de lançamento da missão Apollo 11:", neil["Datum"].values[0])
```

```python
# Esses são os módulos que precisaremos utilizar
import pandas as pd
import requests
```

```python
# A URL a seguir contém o site a ser raspado
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_Human_Development_Index'
# Obtenha os dados do site utilizando a função read_htmĺ() da biblioteca pandas
dfs = pd.read_html(url)
```

```python
# Leia os dados baixados e atribua a tabela que possui os dados do Índice HDI para cada país à uma nova variável tabela
tabela = dfs[1]
tabela
```

```python
# Utilize o operador columns para alterar as colunas para o seguinte valor = ['#','Change since 2015','Nation','HDI','Average annual growth']
tabela.columns = ['#','Change_since_2015','Nation','HDI','Average_annual_growth']
tabela
```

```python
# Quantas colunas, elementos e valores não nulos nossa tabela possui?
tabela.info()
```

```python
# Obtenha a média e desvio-padrão para a tabela
print("Média:\n",tabela.mean(numeric_only=True),"\n")
print("Desvio padrão:\n",tabela.std(numeric_only=True),"\n")
```

```python
# A coluna 'Average annual growth' possui um valor faltante
# Remova o valor de índices [43 130] utilizando a função drop()
tabela_nn = tabela.drop(index=[43,130])
tabela_nn = tabela_nn.reset_index(drop=True)
tabela_nn
```

```python
# A coluna 'Average annual growth' possui dados numéricos, mas não foi possível realizar a análise para essa coluna
# Utilize a função replace() para substituir o símbolo '%' pelo caracter vazio '' e transforme a coluna em número real com a função float()
# Caso tenha dúvida, comece com um elemento e depois passe para a coluna com uma compreensão de lista, como feito em aula
# Utilize o código do script feito em aula para ajudar na escrita do código

tabela_nn["Average_annual_growth"] = [i.replace('%','') for i in tabela_nn["Average_annual_growth"]]
tabela_nn["Average_annual_growth"] = [float(i) for i in tabela_nn["Average_annual_growth"]]
tabela_nn.dtypes
```

```python
# Obtenha a descrição estatística novamente para incluir os resultados da nova coluna numérica
print(tabela_nn.info(),"\n")
print("Média:\n",tabela_nn.mean(numeric_only=True),"\n")
print("Desvio padrão:\n",tabela_nn.std(numeric_only=True),"\n")
```

```python
# Qual é a posição do Brasil na lista?

brasil = tabela_nn[tabela_nn["Nation"] == "Brazil"]
print("Posição do Brasil:",brasil["#"].values[0])
```

```python
def starting_spells(cantrips:list):
    spellbook = pd.DataFrame({"name":[]})
    listed = []
    for spell in cantrips:
        to_add = spells[spells["name"] == spell]
        listed.append(to_add)
        if to_add["name"].values[0] not in spellbook["name"]:
            my_spells = pd.concat(listed, ignore_index=True)
    book = my_spells.to_csv("Spellbook.csv")
    permanent = pd.read_csv("Spellbook.csv")
    return permanent
```

```python
spellbook = starting_spells(["Chill Touch","Ray of Frost", "Shocking Grasp"])
spellbook = spellbook.drop(labels='Unnamed: 0',axis=1)
spellbook
```

# Spells at 1st level

```python
def learn_spells(two_spells:list):
    listed = []
    for spell in two_spells:
        add = spells[spells["name"] == spell]
        listed.append(add)
        new_spells = pd.concat(listed, ignore_index=True)
        learn = pd.concat([spellbook,new_spells],ignore_index=True)
        learned = learn.to_csv("Spellbook.csv")
        ink = pd.read_csv("Spellbook.csv")
    ink = ink.drop(labels='Unnamed: 0',axis=1)
    ink = ink.sort_values("level", ignore_index=True)
    return ink

spellbook = learn_spells(["Mage Armor", "Inflict Wounds", "False Life",
                          "Ray of Sickness", "Ice Knife", "Frost Fingers"])
spellbook
```

# Preparing spells

## Cantrips

```python
def known_cantrips():
    character_level = int(input("Enter your character level: ").strip().lower())
    if character_level < 4:
        three_cantrips = []
        cantrips = input("Enter 3 cantrips to prepare them: ").split(",")
        if len(cantrips) != 3:
            raise ValueError("Only 3 cantrips available at your level.")
        else:
            for cantrip in cantrips:
                cantrip_df = spellbook[spellbook["name"] == cantrip]
                three_cantrips.append(cantrip_df)
            known_cantrips = pd.concat(three_cantrips, ignore_index=True)
            return known_cantrips
    elif character_level >= 4 and character_level < 10:
        four_cantrips = []
        cantrips = input("Enter 4 cantrips to prepare them: ").split(",")
        if len(cantrips) != 4:
            raise ValueError("Only 4 cantrips available at your level.")
        else:
            for cantrip in cantrips:
                cantrip_df = spellbook[spellbook["name"] == cantrip]
                four_cantrips.append(cantrip_df)
            known_cantrips = pd.concat(four_cantrips, ignore_index=True)
            return known_cantrips
    else:
        five_cantrips = []
        cantrips = input("Enter 5 cantrips to prepare them: ").split(",")
        if len(cantrips) != 5:
            raise ValueError("Only 5 cantrips available at your level.")
        else:
            for cantrip in cantrips:
                cantrip_df = spellbook[spellbook["name"] == cantrip]
                five_cantrips.append(cantrip_df)
            known_cantrips = pd.concat(five_cantrips, ignore_index=True)
            return known_cantrips

my_cantrips = known_cantrips()
my_cantrips
```

## Prepared spells

```python
def prepare_spells():
    character_level = int(input("Enter your character level: ").strip().lower())
    character_int = int(input("Enter your Intelligence modifier: ").strip().lower())
    number_available = character_int + character_level
    prepare = input(f"Enter {number_available} spells to prepare them: ").split(",")
    if len(prepare) != number_available:
        raise ValueError(f"Number of spells must be equal to {number_available}.")
    else:
        my_spells = []
        for spell in prepare:
            spell_df = spellbook[spellbook["name"] == spell]
            my_spells.append(spell_df)
        prepared_spells = pd.concat(my_spells, ignore_index=True)
        prepared_spells = prepared_spells.sort_values("level", ignore_index=True)
        return prepared_spells
    
my_prepared_spells = prepare_spells()
my_prepared_spells
```

```python
def required_spell_components():
    my_materials = my_prepared_spells[my_prepared_spells["M"].notna()]
    for name in my_materials["name"]:
        spell = my_materials[my_materials["name"] == name]
        print(f"Spell: {spell['name'].values[0]}\n",
              "Materials:",spell["M"].values[0],"\n")
    return

required_spell_components()
```

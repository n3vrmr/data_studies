```python
# Importando os dados
import pandas as pd

dados = pd.read_csv('ds_salaries.csv')
dados
```

```python
# Informações básicas
## Determine o número e nome das colunas do conjunto de dados
columns = dados.columns
print(columns)
print("Número de colunas:",len(columns))

## Determine o tipo de dado de cada coluna
print("Tipos de dados:",dados.dtypes)

## Determine o número de observações do conjunto de dados
print("Número de observações:",dados.count())

## Determine os valores máximos e mínimos das colunas numéricas
##e a amplitude desses dados
print("Máximos:")
max_values = dados.max(numeric_only=True)
print(max_values)

print("Mínimos:")
min_values = dados.min(numeric_only=True)
print(min_values)

print("Amplitude:")
print("work_year:",max_values['work_year']-min_values['work_year'])
print("salary:",max_values['salary']-min_values['salary'])
print("salary_in_usd:",max_values['salary_in_usd']-min_values['salary_in_usd'])
print("remote_ratio:",max_values['remote_ratio']-min_values['remote_ratio'])
```

```python
# Estatística descritiva
## Obtenha a média e mediana das colunas numéricas
media = dados.mean(numeric_only=True)
mediana = dados.median(numeric_only=True)
print("Média:",media)
print("Mediana:",mediana)
```

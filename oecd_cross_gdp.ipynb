```python
import pandas as pd
gdp = pd.read_excel("WEOApr2022all.xlsx")

def keep_rows(descriptor:str):
    """
    Removes the rows which do not have the same 'Subject Descriptor' as the
    parameter.

    Parameters
    ----------
    descriptor : Subject Descriptor

    Returns
    -------
    gdp_pc: the original dataframe, cleaned from unwanted entries.

    """
    gdp_pc = gdp[gdp['Subject Descriptor'] == descriptor]
    return gdp_pc

gdp = keep_rows('Gross domestic product per capita, current prices')
gdp
```

```python
def keep_usd():
    """
    Removes the rows which are not measured in US Dollars.

    Returns
    -------
    gdp_usd: the original dataframe, containing only the entries measured in
    USD.

    """
    gdp_usd = gdp[gdp['Units'] == 'U.S. dollars']
    return gdp_usd

gdp = keep_usd()
gdp
```

```python
def create_usable_df():
    """
    Creates a new DataFrame object, containing only the country names and their
    corresponding GDP per capita in USD for the year of 2020 (also converts the
    GDP values to float).

    Returns
    -------
    usable: DataFrame object, cleaned for the purposes of Example 1-1 in
    Aurélien Géron's Hands-on Machine Learning.

    """
    usable = pd.DataFrame({"Country":gdp['Country'],
                           "GDP per capita (USD)":gdp[2020]})
    usable["GDP per capita (USD)"] = [float(i) for i in usable["GDP per capita (USD)"]]
    return usable

gdp_final = create_usable_df()
gdp_final
```

```python
import pandas as pd

oecd_bli = pd.read_excel("oecd_bli_excel.xlsx")
oecd_final = pd.DataFrame({"Country":oecd_bli['Country'],
                           "Life satisfaction":oecd_bli['Life satisfaction']})
oecd_final
cyprus = gdp_final[gdp_final['Country'] == 'Cyprus']
cyprus
```

```python
oecd_cross_gdp = pd.merge(gdp_final,oecd_final, on='Country')
oecd_cross_gdp
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import linear_model, neighbors

oecd_cross_gdp = pd.merge(gdp_final,oecd_final, on='Country')

x = np.c_[oecd_cross_gdp['GDP per capita (USD)']]
y = np.c_[oecd_cross_gdp['Life satisfaction']]

oecd_cross_gdp.plot(kind="scatter", x="GDP per capita (USD)", y="Life satisfaction")
plt.show()

lin_reg_model = linear_model.LinearRegression()
lin_reg_model.fit(x,y)
oecd_cross_gdp.corr()
```

```python
# Predict life satisfaction for Cyprus
x_new = [[27783.759]]
print(lin_reg_model.predict(x_new))

clf = neighbors.KNeighborsRegressor(n_neighbors=5)
clf.fit(x,y)
print(clf.predict(x_new))
```
